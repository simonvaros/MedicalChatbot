```python
import pandas as pd
from torchtext.data.metrics import bleu_score
pd.set_option('display.max_colwidth', None)
```

```python
data = pd.read_csv('../../data/test_data_pairs.csv')
```

```python
data.info()
```

```python
data.columns = ['nothing', 'question', 'answer', 'greedy_answer', 'beam_answer', 'sampling_answer']
```

```python
data.head()
```

```python
def compute_bleu(predicted, real):
    candidate_corpus = [predicted.split(' ')]
    references_corpus = [[real.split(' ')]]

    return bleu_score(candidate_corpus, references_corpus)
```

```python
data['greedy_bleu'] = data.apply(lambda x: compute_bleu(x['greedy_answer'], x['answer']), axis=1)
```

```python
data.head(10)
```

```python
data.greedy_answer.values.tolist()
```

```python
from ast import literal_eval
data['answer'] = data['answer'].apply(literal_eval)
```

```python
greedy_answers = data.greedy_answer.values.tolist()
```

```python
candidates = []
for i, answer in enumerate(greedy_answers):
    candidates.append(answer.split(' '))
```

```python
references = []
real_answers = data.answer.values.tolist()

for i, answer in enumerate(real_answers):
    if i == 1: 
        print(answer[0])
    references.append(answer)
```

```python
print(bleu_score(candidates, references))
```

```python
data = pd.read_csv('../../data/test_data_processed.csv')
```

```python
data.info()
```

```python
data = data[['question', 'answer']]
```

```python
data['answer'] = data.groupby(['question'])['answer'].transform(lambda x: ','.join(x))
data = data.drop_duplicates()
```

```python
def split_answer(answer):
    print(answer)
    all_answers = answer.split(',')
    new = []
    
    for idx, a in enumerate(all_answers):
        new.append(a.split(' '))
        
    return new
```

```python
data['answer'] = data.apply(lambda x: split_answer(x['answer']), axis=1)
```
