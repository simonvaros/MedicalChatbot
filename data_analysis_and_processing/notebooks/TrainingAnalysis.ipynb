```python
data = pd.read_csv("./latest_training.csv", delimiter=",")
```

```python
data.plot.line(y=['Training loss', 'Validation loss'], x='Epoch', figsize=(15,10), fontsize=15)
```

```python
def countPerplexity(loss):
    return e**loss
```

```python
data['Training perplexity'] = data.apply(lambda x: countPerplexity(x['Training loss']), axis=1)
data['Validation perplexity'] = data.apply(lambda x: countPerplexity(x['Validation loss']), axis=1)
```

```python
data.plot.line(y=['Training perplexity', 'Validation perplexity'], x='Epoch', figsize=(15,10), fontsize=15)
```
