```python
import pandas as pd
from math import e
```

```python
data = pd.read_csv('../../priebeh_trenovania.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
def countPerplexity(loss):
    return e**loss
```

```python
data['perplexity'] = data.apply(lambda x: countPerplexity(x['loss']), axis=1)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.plot.line(x='iteration', y='loss', figsize=(15,10))
```
